```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file_path = 'data.xls'
```

```python
data = pd.read_csv(open(file_path))
data.columns = data.columns.str.lower()
```

```python
data.head()
```

```python
numerical_cols = [col for col in data.columns if np.issubsctype(data[col],np.number)]
```

```python
categorical_cols = [col for col in data.columns if data[col].dtypes =='O']
```

```python
import datetime as dt
import re
```

```python
date_cols = []
for col in numerical_cols:
    match = re.search(r'\d{4}\d{2}',col)
    date_cols.append(dt.datetime.strptime(match.group(), '%Y%m').date())
numerical_cols = date_cols
```

```python
data.columns = list(data.columns[:3]) + date_cols 
```

```python
data.head()
```

What is the distribution of major crime categories? 
</div>

In order to find the distribution of the major crime categories, first, we need to find what the major crime category is.

```python
g_crime_cat = data.groupby('majortext')
```

```python
major_crimes = g_crime_cat.sum().sum(axis=1).reset_index()
major_crimes.columns = ['majortext','n_crims']
```

```python
major_crimes
```

<div class='alert alert-block alert-warning' style='font-size:150%'>

### Bar Plot

Using bar plot to illusterate the distribution of major crime categories.

</div>

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.bar(major_crimes['majortext'],major_crimes['n_crims'])
ax.set_title('Distribution of major crimes')
ax.set_xlabel('Major crimes categories')
ax.tick_params(axis='x', rotation=70)
ax.set_ylabel('n_crimes')
```

<div class='alert alert-block alert-warning' style='font-size:150%'>

```python
print('Major crime category is:' + major_crimes.argmax())
```

Now that we find that the major crime category is theft, let's slice the data with this category.

```python
Theft_data = data[data['majortext']=='Theft']
```

```python
n,bins,patches = plt.hist(Theft_data[numerical_cols].values.ravel(),50,density=True,facecolor='g')
plt.xlabel('n_crimes')
plt.ylabel('probability')
plt.title('Histogram of number of crimes for Theft category')
```

### Bubble plot x_axis: "major categories",  y_axis: "Region"

</div>

```python
g_region_major = data.groupby(['majortext','lookup_boroughname']).sum().sum(axis=1).reset_index()
g_region_major.columns = ['majortext','lookup_boroughname','count']
```

```python
fix, ax = plt.subplots(figsize=(20,15))
plt.scatter(g_region_major['majortext'],g_region_major['lookup_boroughname'], s=g_region_major['count']/20,cmap="Blues", alpha=0.8, edgecolors="grey", linewidth=2)
plt.xticks(rotation=80)
plt.grid(True)
```
